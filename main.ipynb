```python
def object_to_int(data,coloum):
    index = -1
    info_dict = {}
    all_info = []
    for i in data[coloum]:
        if i not in info_dict:
            index = index + 1
            info_dict[i] = index
    for x in data[coloum]:
        all_info.append(info_dict[x])
    return all_info,info_dict
```

```python
data.rename({'Location':'City'},axis=1,inplace=True)
```
